```python
import os
import pandas as pd
```

```python
# File path
file_path = '../Data/Fireline_5mPoints_240205_Drops.csv'

# Read the CSV file
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
# Regular expression pattern to extract percentage
pattern = r'(\d+)%'

# Extract and convert the percentage values; fill non-matching entries with 0
df['EVC Percentage'] = df['EVC_CLASSNAMES'].str.extract(pattern).fillna(0).astype(int)

# Display the DataFrame to verify the results
df.head()
```

```python
# Regular expression pattern to extract height in meters (handles integers and decimals)
pattern = r'(\d+\.?\d*)\s*meters?'

# Extract and convert the height values; fill non-matching entries with 0
df['Height_Meters'] = df['EVH_CLASSNAMES'].str.extract(pattern).fillna(0).astype(float)

# Display the DataFrame to verify the results
df.head()
```

```python
# Number of parts to divide the file into
num_parts = 10

# Calculate the number of rows per part
rows_per_part = len(df) // num_parts

# Ensure the 'Data' folder exists
output_folder = 'Data'
os.makedirs(output_folder, exist_ok=True)

# Split and save the files
for part in range(num_parts):
    start_row = part * rows_per_part
    # For the last part, include all remaining rows
    end_row = (part + 1) * rows_per_part if part < num_parts - 1 else len(df)
    df_part = df.iloc[start_row:end_row]
    output_file = os.path.join(output_folder, f'Fireline_5mPoints_part_{part+1}.csv')
    df_part.to_csv(output_file, index=False)

print(f'Files are split and saved in {output_folder}')
```
